```python
import pandas as pd
url="https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data2.csv"
data=pd.read_csv(url)
data
```

```python
#Label encoding
data["EducationLevel"]=data["EducationLevel"].map({"高中以下":0, "大學":1, "碩士以上":2})
data
```

```python
#將城市ABC轉換成三個特徵值
from sklearn.preprocessing import OneHotEncoder
onehot_encoder=OneHotEncoder()
onehot_encoder.fit(data[["City"]])   #要二維矩陣
city_encoded=onehot_encoder.transform(data[["City"]]).toarray()
city_encoded

```

```python
data[["CityA","CityB","CityC"]]=city_encoded
data
```

```python
data=data.drop(["City","CityC"],axis=1)  #刪掉column axis=1
data
```

```python
from sklearn.model_selection import train_test_split   #訓練集&測試集分開
x=data[["YearsExperience","EducationLevel","CityA","CityB"]]
y=data["Salary"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=86)
x_train=x_train.to_numpy()
x_test=x_test.to_numpy()
```

```python
#特徵縮放 因為Years experience這欄位的數值顯著大於其他 因此要透過標準化(x-mean(x))/std(x)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_test
```

```python
import numpy as np
w=np.array([1,2,3,4])  #假設w1,w2,w3,w4
b=1
y_pred=(x_train*w).sum(axis=1)+b  #計算w1*x1+w2*x2+w3*x3+w4*x4...  sum(axis=1)代表row總和
#月薪y=w1*年資+w2*學歷+w3*cityA+w4*cityB+b
```

```python
((y_train-y_pred)**2).mean()   #設定cost function=(真實值-預測值)**2 作為評判標準
def compute_cost(x,y,w,b):
  y_pred=(x*w).sum(axis=1)+b
  cost=((y-y_pred)**2).mean()
  return cost
w=np.array([0,2,2,4])
b=1
compute_cost(x_train,y_train,w,b)
```

```python
y_pred=(x_train*w).sum(axis=1)+b
b_gradient=(y_pred-y_train).mean()
w_gradient=np.zeros(x_train.shape[1])  #用shape來判斷欄位數 有w1,w2,w3,w4
for i in range(x_train.shape[1]):
  w_gradient[i]=(x_train[:,i]*(y_pred-y_train)).mean()
w_gradient,b_gradient
# w_gradient
# x_train.shape
# w1_gradient=(x_train[:,0]*(y_pred-y_train)).mean()     #cost function對w1微分,w1方向的斜率
# x1= x_train[:,0]
# w2_gradient=(x_train[:,1]*(y_pred-y_train)).mean()
# w3_gradient=(x_train[:,2]*(y_pred-y_train)).mean()
# w4_gradient=(x_train[:,3]*(y_pred-y_train)).mean()
# w1_gradient,w2_gradient,w3_gradient,w4_gradient
```

```python
def compute_gradient(x,y,w,b):
  y_pred=(x*w).sum(axis=1)+b
  w_gradient=np.zeros(x.shape[1])
  b_gradient=(y_pred-y).mean()
  for i in range(x.shape[1]):
    w_gradient[i]=(x[:,i]*(y_pred-y)).mean()
  return w_gradient, b_gradient   #計算出w,b的斜率
```

```python
w=np.array([0,2,2,4])
b=1
compute_gradient(x_train,y_train,w,b)
```

```python
w=np.array([1,2,2,4])
b=1
learning_rate=0.001
w_gradient,b_gradient=compute_gradient(x_train,y_train,w,b)
print(compute_cost(x_train,y_train,w,b))
w=w-w_gradient*learning_rate
b=b-b_gradient*learning_rate
w,b
print(compute_cost(x_train,y_train,w,b))
```

```python
np.set_printoptions(formatter={"float": "{: .2e}".format})
def gradient_descent(x,y,w_init,b_init,learning_rate,cost_function,gradient_function,run_iter,p_iter=1000):
  w=w_init
  b=b_init
  c_hist=[]
  w_hist=[]
  b_hist=[]
  for i in range(run_iter):
    w_gradient, b_gradient=gradient_function(x,y,w,b)
    w=w-w_gradient*learning_rate
    b=b-b_gradient*learning_rate
    cost=cost_function(x,y,w,b)
    w_hist.append(w)
    c_hist.append(cost)
    b_hist.append(b)
    if i%p_iter==0:
      print(f"Iteration{i:5}: Cost: {cost }, W: {w }, B: {b }, w_gradient: {w_gradient }, b_gradient: {b_gradient } ")
  return w,b,w_hist,b_hist,c_hist
```

```python
w_init=np.array([1,2,2,4])
b_init=0
learning_rate=1.0e-2
run_iter=20000

w_final,b_final,w_hist,b_hist,c_hist=gradient_descent(x_train,y_train,w_init,b_init,learning_rate,compute_cost,compute_gradient,run_iter,p_iter=1000)
```

```python
w_init=np.array([1,2,2,4])
b_init=0
learning_rate=1.0e-2
run_iter=20000

w_final,b_final,w_hist,b_hist,c_hist=gradient_descent(x_train,y_train,w_init,b_init,learning_rate,compute_cost,compute_gradient,run_iter,p_iter=1000)
```
